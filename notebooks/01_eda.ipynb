# Automobile Insurance Fraud - EDA

This notebook focuses on **exploratory data analysis (EDA)** for the fraud detection dataset.

- Input: `data/Automobile_insurance_fraud_cleaned.csv`
- Output: EDA insights and plots

Model training is separated into: `notebooks/02_modeling.ipynb`.


```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from insurance_project.features import add_age_and_tenure_groups
from insurance_project.paths import data_path

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

df = pd.read_csv(data_path("Automobile_insurance_fraud_cleaned.csv"))
df = add_age_and_tenure_groups(df)
df.shape
```

## 1) Quick look

- Dataset shape + a few rows
- Target distribution (`fraud_reported`)


```python
display(df.head())
display(df["fraud_reported"].value_counts(dropna=False))
```

## 2) Target vs key features

A few quick plots to see relationships between the target and important features.


```python
sns.countplot(x="incident_severity", hue="fraud_reported", data=df)
plt.title("incident_severity vs fraud_reported")
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="insured_hobbies", hue="fraud_reported", data=df)
plt.title("insured_hobbies vs fraud_reported")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x="fraud_reported", y="total_claim_amount", data=df)
plt.title("total_claim_amount by fraud_reported")
plt.show()
```

## 3) Feature engineering (lightweight)

`age_group` and `tenure_group` make EDA easier to read (and will also be used for one-hot encoding later).


```python
# The grouping columns were added in the setup cell via `add_age_and_tenure_groups`.
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.barplot(x="age_group", y="fraud_reported", data=df, errorbar=None, palette="Reds")
plt.title("Fraud rates by age group")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(x="tenure_group", y="fraud_reported", data=df, errorbar=None, palette="Blues")
plt.title("Fraud rates by tenure of customer")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
